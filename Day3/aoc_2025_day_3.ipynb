```python
# READ INPUT

file_path = 'input.txt'
lines_raw = []
lines = []

with open(file_path, 'r') as file:
    lines_raw = file.readlines()

for line in lines_raw:
    lines.append(line.strip())

# Each element in 'lines' contains a bank
# Execute for each bank one by one
# Search for x-largest
# Last x positions should be ignored
# Repeat 12 times

joltages = []
total_digits = 12

for bank in lines:

    # Resets for bank
    digits = total_digits            # current digit on the 12-digit number
    batteries = [False] * len(bank)  # bool for each battery if already chosen
    battery_value = []               # battery values for the 12-digit number
    previous_idx = -1                # to ensure only batteries on the right of the last idx can be chosen
    
    for digit_idx in range(total_digits):
        
        # Resets for digit
        digits -= 1         # to calculate how many to ignore at the end
        largest = 0         # to safe battery value
        largest_idx = 0     # to track battery position
    
        # Iterate through each battery
        for battery_idx in range(len(bank)):
            battery = int(bank[battery_idx])

            if (batteries[battery_idx] == False       # not chosen yet
               and battery_idx > previous_idx         # right of previous digit
               and battery_idx < len(bank) - digits   # not at the right edge
               and battery > largest):                # new largest
                largest = battery
                largest_idx = battery_idx

        # Safe largest info
        batteries[largest_idx] = True
        battery_value.append(largest)
        previous_idx = largest_idx

    # Calculate solution
    joltage = 0
    for idx in range(total_digits):
        joltage += battery_value[idx] * (10 ** (total_digits - 1 - idx))
    joltages.append(joltage)

print(sum(joltages))
```
