```python
# READ INPUT

file_path = 'input.txt'
lines_raw = []
lines = []

with open(file_path, 'r') as file:
    lines_raw = file.readlines()

for line in lines_raw:
    lines.append(line.strip())

# Lines contains rows
# Each row is a string with . or @ where the position equals the column

i_max = len(lines)
j_max = len(lines[0])

# Create array to hold paper counts
roll_counts = []
for i in range(i_max):
    roll_counts.append([0] * j_max)

# Counter for accessable paper rolls
total_access = 0

# Flag to abort once no paper rolls were removed
papers_removed = True

while papers_removed:

    papers_removed = False
    
    # Count '@' symbols
    for i in range(i_max):
        for j in range(j_max):
            symbol = lines[i][j]
    
            # Only consider roll of paper positions
            if symbol != '@':
                continue
    
            paper_count = 0
            
            if i > 0 and j > 0 and lines[i-1][j-1] == '@':
                paper_count += 1
    
            if j > 0 and lines[i][j-1] == '@':
                paper_count += 1
    
            if i < i_max - 1 and j > 0 and lines[i+1][j-1] == '@':
                paper_count += 1
    
            if i > 0 and lines[i-1][j] == '@':
                paper_count += 1
    
            if i < i_max - 1 and lines[i+1][j] == '@':
                paper_count += 1
    
            if i > 0 and j < j_max - 1 and lines[i-1][j+1] == '@':
                paper_count += 1
    
            if j < j_max - 1 and lines[i][j+1] == '@':
                paper_count += 1
    
            if i < i_max - 1 and j < j_max - 1 and lines[i+1][j+1] == '@':
                paper_count += 1
    
            if paper_count < 4:
                total_access += 1
                papers_removed = True
            
            roll_counts[i][j] = paper_count

    # Remove '@' symbols with total_access < 4
    for i in range(i_max):
        for j in range(j_max):
            if roll_counts[i][j] < 4:
                lines[i] = lines[i][:j] + '.' + lines[i][j + 1:]
                roll_counts[i][j] = 0

print(total_access)
```
