```python
# READ INPUT

file_path = 'input.txt'
lines_raw = []
lines = []

with open(file_path, 'r') as file:
    lines_raw = file.readlines()

for line in lines_raw:
    lines.append(line)

vertical_lines = {}
for i in range(len(lines)):
    line = lines[i]
    for j in range(len(line)):
        if i == 0:
            vertical_lines[j] = []
        v_array = vertical_lines[j]
        v_array.append(line[j])

# vertical_lines contains all vertical numbers and symbol in each column with column_index as the key

vertical_calculations = {}
v_calc_idx = 0
last_symbol = "-"

for i in range(len(vertical_lines)):
    symbol = vertical_lines[i][-1]
    
    if symbol == '*' or symbol == '+':
        last_symbol = symbol
        v_line_dict = {}
        v_line_dict[symbol] = []
        vertical_calculations[v_calc_idx] = v_line_dict
        v_calc_idx += 1

    number = ""
    for j in range(len(vertical_lines[i]) - 1):
        digit = str(vertical_lines[i][j])
        if digit != " ":
            number += digit

    if len(number.strip()) > 0:
        v_line_dict[last_symbol].append(number)

results = []
for calculation in vertical_calculations.values():
    symbol = list(calculation.keys())[0]
    numbers = list(calculation.values())[0]

    if symbol == '*':
        result = 1
        for number in numbers:
            result *= int(number)
        results.append(result)
    
    if symbol == '+':
        result = 0
        for number in numbers:
            result += int(number)
        results.append(result)

print(sum(results))
```
