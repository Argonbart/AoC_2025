```python
# READ INPUT

file_path = 'input.txt'
lines_raw = []
lines = []

with open(file_path, 'r') as file:
    lines_raw = file.readlines()

for line in lines_raw:
    lines.append(line.strip())

# Each element in 'lines' contains a rotation
# Execute rotations one by one

current_number = 50
zero_counter = 0

for line in lines:
    rotation_dir = line[0]
    rotation_amount = int(line[1:])
    rotation_left = rotation_dir == "L"

    # increase zero counter for big spins
    # set rotation_amount to 0-100
    if rotation_amount > 99:
        zero_counter += rotation_amount // 100
        rotation_amount = rotation_amount % 100

    # save before and after values
    # rotate number
    current_number_before = current_number
    current_number += rotation_amount * (-1) ** int(rotation_left)
    current_number = (current_number + 100) % 100
    current_number_after = current_number

    # check if landed on zero or zero was crossed
    if (rotation_left and current_number_before != 0 and (current_number_after > current_number_before)) or (not rotation_left and current_number_before != 0 and (current_number_after < current_number_before)) or current_number == 0:
        zero_counter += 1

print(zero_counter)
```
