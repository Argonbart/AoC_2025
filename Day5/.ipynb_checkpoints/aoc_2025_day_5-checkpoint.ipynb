```python
# READ INPUT

file_path = 'input.txt'
starts = []
ends = []

with open(file_path, 'r') as file:
    line = file.readline().strip()
    while line != "":
        new_range = line.split("-")
        # if new_range[0] == new_range[1]:
        #     print(new_range[0], new_range[1])
        starts.append(int(new_range[0]))
        ends.append(int(new_range[1]))
        line = file.readline().strip()

# Ranges contain pair-wise ranges with starts[i] and ends[i] being a pair

starts.sort()
ends.sort()

current_range_start_id = 0
inside_range_counter = 0
fresh_ids_count = 0

while starts or ends:

    if not starts:
        fresh_ids_count += (ends[-1] - current_range_start_id + 1)
        break

    if starts[0] == ends[0]:
        if inside_range_counter == 0:
            if len(starts) > 1:
                if starts[1] != starts[0]:
                    fresh_ids_count += 1
            else:
                fresh_ids_count += 1
        starts.pop(0)
        ends.pop(0)
        continue

    if starts[0] < ends[0]:
        if inside_range_counter == 0:
            current_range_start_id = starts[0]
        inside_range_counter += 1
        starts.pop(0)
        continue

    if ends[0] < starts[0]:
        inside_range_counter -= 1
        if inside_range_counter == 0:
            fresh_ids_count += (ends[0] - current_range_start_id + 1)
        ends.pop(0)
        continue

print(fresh_ids_count) # 49 56 61
```
