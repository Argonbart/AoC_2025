```python
def check_number(number):
    number = str(number)

    # Test different sequence lengths
    for sequence_len in range(1, len(number)):
        
        # Only consider sequence length that fits in the total number length evenly
        if (len(number) % sequence_len) != 0:
            continue

        # Set first sequence based on length
        original_sequence = number[:sequence_len]

        # Iterate through sequences and check if it fits the original sequence
        found_invalid = True
        for sequence_idx in range(len(number) // sequence_len):
            current_sequence = number[sequence_idx * sequence_len : (sequence_idx + 1) * sequence_len]

            # Abort if any did not fit
            if current_sequence != original_sequence:
                found_invalid = False
                break

        # Return true if a sequence length led to an invalid id
        if found_invalid:
            return True
    
    return False
```

```python
# READ INPUT

file_path = 'input.txt'
ranges = []

with open(file_path, 'r') as file:
    ranges = file.readline().split(",")

# Each element in 'ranges' contains a range
# Execute range checks one by one

invalid_ids = []
invalid_id_counter = 0

for range_item in ranges:
    edges = range_item.split("-")
    left = int(edges[0])
    right = int(edges[1])
    current_number = left
    while current_number <= right:
        if check_number(current_number):
            invalid_ids.append(current_number)
            invalid_id_counter += 1
        current_number += 1

print(sum(invalid_ids))
```
